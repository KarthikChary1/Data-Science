```python
from autogluon.timeseries import TimeSeriesPredictor, TimeSeriesDataFrame
import yfinance
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


df = yfinance.download("^NSEI",interval = "1d")
df.reset_index(inplace=True)


df.columns = [i[0] for i in df.columns]

df.rename(columns = {"Datetime":"timestamp","Date":"timestamp","Close":"target"},inplace = True)
dataset = df[["timestamp","target"]]

dataset["timestamp"] = dataset["timestamp"].dt.strftime("%d-%m-%Y %H:%M:%S")
dataset["timestamp"] = pd.to_datetime(dataset["timestamp"],format="%d-%m-%Y %H:%M:%S")

dataset['item_id'] = 1

training = dataset.iloc[:-10,:]
testing = dataset.iloc[-10:,:]

df = TimeSeriesDataFrame.from_data_frame(training)

predictor = TimeSeriesPredictor(prediction_length=14,freq="1d",).fit(
    df,
    hyperparameters={
        "Chronos": {"model_path": "autogluon/chronos-bolt-base"},
    },
)

predictions = predictor.predict(df)

predictions.reset_index(inplace=True)
predictions["timestamp"]=pd.to_datetime(predictions["timestamp"])
predictions["day"] = predictions["timestamp"].dt.day_name()
predictions = predictions[(predictions["day"]!="Sunday") & (predictions["day"]!="Saturday")]


```
